```python
def convert_temperarture(temp,unit):
    """"This changes temp b/w celsius and fahrenheit"""
    if unit=="C":
        return temp*9/5+32
    elif unit=="F":
        return (temp-32)*5/9
    else:
        return None
print(convert_temperarture(25,'C'))
print(convert_temperarture(100,'F'))
    
```

```python
def is_palindrome(text):
    text=text.lower().replace(" ","")# removes the empty space to no space
    if text[::-1]==text:
        return True
    return False
is_palindrome("H elleh")
```

```python
#Factorial using recursion
def factorial(num):
    if num==0:
        return 1
    else:
        return num*factorial(num-1)
factorial(6)
```
