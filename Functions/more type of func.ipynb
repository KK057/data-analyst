```python
def get_name(person):
    return person['name']

people = [
    {'name': 'Krish', 'age': 32},
    {'name': 'Jack', 'age': 33}
]

list(map(get_name, people))
```

```python
#filter func
#it gives the iterables for which the func is true
def even(num):
    if num%2==0:
        return True

lst=[1,2,3,45,6,78,94,5,32,21]
print(list(map(even,lst)))
print(list(filter(even,lst)))
```

```python
#with lambda
greatr_5=list(filter(lambda num:num>5,lst))
print(greatr_5)
```

```python
#multiple condition
output=list(filter(lambda x:x>5 and x%2==0, lst))
print(output)
```

```python
#to check age is greater than 18
people = [
    {'name': 'Krish', 'age': 30},
    {'name': 'Jack', 'age': 3},
    {'name': 'rohan', 'age': 19}
]
def age_grt_18(people):
    return people['age']>18

list(filter(age_grt_18,people))
```
