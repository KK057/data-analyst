```python
#default parameters
def greet(name='newuser'):
    print(f"Hello {name}. welcome here!")
greet("Kaushik")
greet()
```

```python
#Positional arguments
def print_num(*args):
    for values in args:
        print(values)
print_num(48,782,9,6,9,'subhaba')
```

```python
#keywords arguments
def print_details(**args):
    for key,value in args.items():
        print(f"{key}:{value}")
print_details(name='kaushik', age=84,caste='yadav')
```
