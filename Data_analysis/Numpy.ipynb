```python
import numpy as np
```

```python
#creating an array
arr1=np.array([1,22,57,33])#1D array
print(arr1)
print(type(arr1))
print(arr1.shape)
```

```python
print(arr1.reshape(1,4))#1 row and 4 columns
```

```python
arr2=np.array([[5,19,91,859,1,9]])
arr2.shape
```

```python
print("arr1:", arr1)
print("  shape:", arr1.shape)
print("  ndim:", arr1.ndim)
print("  size:", arr1.size)
print("  dtype:", arr1.dtype)
print("  itemsize:", arr1.itemsize)
print("  nbytes:", arr1.nbytes)
print("  strides:", arr1.strides)

print("\narr2:", arr2)
print("  shape:", arr2.shape)
print("  ndim:", arr2.ndim)
print("  size:", arr2.size)
print("  dtype:", arr2.dtype)
print("  itemsize:", arr2.itemsize)
print("  nbytes:", arr2.nbytes)
print("  strides:", arr2.strides)
```

```python
#Numpy vectorized operation
arr2=np.array([48,5,851,84])
print("addition",arr1+arr2)
print("sub",arr1-arr2)
print("mul",arr1*arr2)
print("div",arr1/arr2)
```

```python
#Universal func
print(np.sin(arr1)) #sine
print(np.log(arr1)) #logrithm
print(np.sqrt(arr1))
print(np.exp(arr1))
```

```python
#Array slicing and Indexing
arr=np.array([[65,1,8,9],[19,4,19,91],[4,1,9,4]])
print(f"{arr.shape} \n")
print(arr)
print("\n",arr[0][2]) #1st row and 3rd element
print("\n",arr[1:])#2nd to last row
print("\n",arr[1:,2:])
print("\n",arr[1:,1:3])
```

```python
#Modify the elements
arr[1:,2:3]=100
print(arr)
```

```python
# Statistical concepts using numpy

print("Statistics for arr1:", arr1)
print("Mean:", np.mean(arr1))
print("Standard Deviation:", np.std(arr1))
print("Variance:", np.var(arr1))
print("Min:", np.min(arr1))
print("Max:", np.max(arr1))
```
