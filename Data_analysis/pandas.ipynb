```python
import pandas as pd
```

```python
#Series:1D labeled array,,,
data=[45,8,4,12,7,80,32]
series=pd.Series(data)
print(series,"\n")
```

```python
#From Dictionaries
data1={'box':3,'sun':23,"moon":"Bright"}
series_dict=pd.Series(data1)
print(series_dict,"\n")
```

```python
#Dataframes
#2D labeled table
data={
    'Name':['krish','kaushik','rohit'],
    'Age':[84,10,48],
    'City':['Banglore','Mumbai','Kolkata']
}
df=pd.DataFrame(data)
print(df)
print(type(df))
```

```python
import numpy as np
data_np=np.array(df)
print(data_np)
```
