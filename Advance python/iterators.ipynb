```python
#iterator is an object that lets you traverse through a sequence of values one at a time, without needing to store the entire sequence in memory at once.
num=[854,5,1,48,51,48,611,49]
iterator=iter(num)
print(type(iterator))
```

```python
next(iterator)
```

```python
try:
    print(next(iterator))
except StopIteration:
    print("There are no elements left")
```

```python
def read_data():
    for i in range(5):
        yield i

def square(data):
    for x in data:
        yield x * x

pipeline = square(read_data())

for value in pipeline:
    print(value)
```
