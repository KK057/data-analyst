```python
try:
    num=int(input("Enter a number:"))
    result=687/num
except ValueError:
    print("enter a valid number")
except Exception as ex:# parent exception for all 
    print(ex)
else:#executed when exception didn't hit
    print(f"The result is {result}")
finally:#executed everytimr
    print("Execution completed")
```
