```python
# Magic Methods in Python OOP

class Person:
    def __init__(self, name, age):
        """Constructor - initializes the object"""
        self.name = name
        self.age = age
    
    def __str__(self):
        """Returns a user-friendly string representation"""
        return f"{self.name} is {self.age} years old"
    
    def __repr__(self):
        """Returns a developer-friendly string representation"""
        return f"Person('{self.name}', {self.age})"
    
    def __len__(self):
        """Returns length - in this case, age"""
        return self.age
    
    def __add__(self, other):
        """Overloads + operator"""
        return self.age + other.age
    
    def __lt__(self, other):
        """Overloads < operator"""
        return self.age < other.age
    
    def __eq__(self, other):
        """Overloads == operator"""
        return self.name == other.name and self.age == other.age
    
    def __call__(self, greeting):
        """Makes object callable"""
        return f"{greeting}, I'm {self.name}"

# Test the magic methods
p1 = Person("Alice", 25)
p2 = Person("Bob", 30)

print(str(p1))           # Uses __str__
print(repr(p1))          # Uses __repr__
print(len(p1))           # Uses __len__
print(p1 + p2)           # Uses __add__
print(p1 < p2)           # Uses __lt__
print(p1 == Person("Alice", 25))  # Uses __eq__
print(p1("Hello"))       # Uses __call__
```
