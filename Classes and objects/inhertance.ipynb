```python
#single inheritance
class game:
    def __init__(self,player,game_type):
        self.player=player
        self.game_type=game_type
    def play(self):
        print(f"{self.player} has entered the game")
        
```

```python
class Contra(game):
    def __init__(self,player,game_type,level,mode):
        super().__init__(player,game_type)#this will get from parent class
        self.mode=mode
        self.level=level

    def game_mode(self):
        print(f"The toughness mode of the game is:{self.mode}")
    def game_level(self):
        print(f"Contra game level:{self.level}")
```

```python
contra1 = Contra("kaushik", "action", 5, "normal")
print(contra1.player)
print(contra1.game_mode())
print(contra1.game_level())
```

```python
#Multiple Inheritance

#Base class 1
class Animal:
    def __init__(self,name):
        self.name=name
    def speak(self):
        print("Subclass must implement this method")
#Base class 2
class Pet:
    def __init__(self,owner):
        self.owner=owner

#Derived class
class Dog(Animal,Pet):
    def __init__(self,name,owner):
        Animal.__init__(self,name)
        Pet.__init__(self,owner)
    def speak(self):
        return f"{self.name} says Woof"
```

```python
#Create an object
dog1=Dog("Alexa","Kaushik")
print(dog1.speak())
print(dog1.owner)
```
