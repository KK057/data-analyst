```python
student={"name":"Krish","age":43,"grade":'A'}
print(student)
```

```python
print(student['age'])
#accesssing using get() method
print(student.get('name'))
print(student.get('last_name'))
print(student.get('last_name',"Not available"))#default value will be printed if it is not present
```

```python
#Modfying dictionary Elements
student["age"]=33
print(student)
student["address"]="kolkata"
print(student)
```

```python
del student['name']
print(student)
```

```python
#Dictionary methods
keys=student.keys()
print(keys)
values=student.values()
print(values)
items=student.items()
print(items)#will give list of tuples
```

```python
#shallow copy
#shallow copy avoids the same upadation in both dict
student_copy=student.copy()
print(student_copy)
print(student)
```

```python
student["DOB"]="january"
print(student)
print(student_copy)
```

```python
#iteration
for keys in student.keys():
    print(keys)

for values in student.values():
    print(values)
```

```python
for key,value in student.items():
    print(f"{key}:{value}")
```

```python
#Nested dictionary
students={
    "student1":{'name':'kaushik','age':'84','class':'5'},
    "student2":{'name':'peter','age':'54','class':'8'},
    "student3":{'name':'Osho','age':'35','class':'10'},
    "student4":{'name':'rama','age':'10','class':'9'}
}
print(students)
print(students["student3"]["age"])
```

```python
#iteration
for student_id,student_info in students.items():
    for key,value in student_info.items():
        print(f"{key}:{value}")
```

```python
#dictionary comrehension
even_square={a:a**2 for a in range(20) if a%2==0}
print(even_square)
```

```python
# To count the frequency of numbers in list
lst=[1,2,3,55,4,5,5,3,6,5,3,7,3,68,9,56,8,4,32,3,7]
frequency={}
for num in lst:
    if num in frequency:
        frequency[num]+=1
    else:
        frequency[num]=1
print(frequency)
```
