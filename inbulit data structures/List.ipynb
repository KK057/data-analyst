```python
#modifying list
fruits=[1,2,3]
fruits[0]="watermelon"
print(fruits)
```

```python
fruits.append("orange")
print(fruits)
```

```python
fruits.insert(1,"Apple")
print(fruits)
```

```python
popped_fruits=fruits.pop()#remove and return the last element
print(popped_fruits)
print(fruits)
```

```python
index=fruits.index("Apple")
print(index)
```

```python
fruits.insert(3,"grapes")
print(fruits)
fruits.insert(3,"grapes")
print(fruits)
print(fruits.count("grapes"))
```

```python
#slicing list
numbers=[13,4,57,7,4,64,332,56,88,90]
print(numbers[::-1])#for reversing
print(numbers[::-2])
```

```python
#iteration in list
for index,number in enumerate(numbers):
    print(index,number)
```

```python
l=[]
for x in range(10):
    l.append(x**x)
print(l)
```

```python
#list comprehension
[x**2 for x in range(10)] #at first we have to write the desired output then loop ondition
```

```python
#with condition
[num for num in range(20+1) if num%2==0]
```

```python
#nested
lst1=[1,2,3]
lst2=["a","b","c"]
pair=[(i,j) for i in lst1 for j in lst2]
print(pair)
```

```python
#word length
words=["bookish","downtown","book","pie","BrokenPipeError"]
length=[len(word) for word in words]
print(length)
```
