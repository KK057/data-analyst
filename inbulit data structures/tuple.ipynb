```python
#tuples are ordered collection which are immutable
emptytuple=()
empty=tuple()
print(type(emptytuple))
```

```python
#conversion
numbers=tuple([1,2,3,45,6,78])
numbers
```

```python
#accessing will be similar to list and even slicing
numbers[::1]
```

```python
numbers[1]=33#immutable
```

```python
#tuple methods
print(numbers.count(7))
print(numbers.index(3))
```

```python
#packing and unpacking a tuple
packed_tup=1,"longrange",3.12
print(packed_tup)
a,b,c=packed_tup
print(a)
print(b)
print(c)
```

```python
#unpacking with *
num=(1,2,3,45,7,7,8,75,3,3,9)
first,*middle,last=num
print(first)
print(middle)
print(last)
```
