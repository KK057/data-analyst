```python
#in sets , elements don't follow any order and do not have duplicate elements
my_set={1,3,4,46}
print(my_set)
empty_set=set()
type(empty_set)
```

```python
#basic operations
my_set.add(55)
my_set
```

```python
my_set.remove(3)
print(my_set)
```

```python
my_set.discard(3)#if element is not present, no error will be shown
print(my_set)
```

```python
my_set.clear()
```

```python
#set membership test
my_set={2,3,5,70,78,7,64,20}
print(29 in my_set)
print(70 in my_set)
```

```python
#mathematical operation
set1={1,2,3,4,5,6,81}
set2={6,84,8,2,79,1,2,7,5}
union_set=set1.union(set2)
print(union_set)
intersection_set=set1.intersection(set2)
print(intersection_set)
set1.intersection_update(set2)
print(set1)
```

```python
set1={1,2,3,4,5,6,81}
set2={6,84,8,2,79,1,2,7,5}
print(set1.difference(set2))
print(set2.difference(set1))
set1.difference_update(set2)
print(set1)
```

```python
set1={1,2,3,4,5,6,81}
set2={6,84,8,2,79,1,2,7,5}
set1.symmetric_difference(set2)#unique elements at both will be removed
```

```python
#sets method
set1={1,2,3,4,5,6,81}
set2={6,84,8,2,79,1,2,7,5}
print(set1.issubset(set2))
print(set1.issuperset(set2))
```
